```
def sambda(x):
  s=[]
  for j, y in enumerate(x['cases']):
    if j == 0:
      s.append(0)
      h = y
    else:
      z = y - h
      s.append(z)
      h = y
  return s


```

```
def weekaverage(x):
  'weekly average growth rates'
  sometin = []
  sometinelse = []
  for y in x:
    if len(sometin) == 7:
      z = np.mean(sometin)
      sometinelse.append(y)
      sometin = []
      sometin.append(y)
    else:
      sometin.append(y)
  return sometinelse
```

```
def growth(x):
  alacart = []
  'establishes weekly average growth rates befoer and after'
  for y, z in enumerate(x['cases']):
    if y == 0:
      alacart.append(0)
      h = z
    else:
      a = (np.log(z) - np.log(h)) *100
      alacart.append(a)
      h = z
  averagesperweek = weekaverage(alacart)
  return averagesperweek
```

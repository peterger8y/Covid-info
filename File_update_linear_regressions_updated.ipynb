```
def growth1(x):
  alacart1 = []
  'establishes the day by day growth rates'
  for y, z in enumerate(x['cases']):
    if y == 0:
      alacart1.append(0)
      h = z
    else:
      a = (np.log(z) - np.log(h)) *100
      alacart1.append(a)
      h = z
  return alacart1
  
```

```
def weekaverage(x):
  'weekly average growth rates'
  sometin = []
  sometinelse = []
  for y in x:
    if len(sometin) == 7:
      z = np.mean(sometin)
      sometinelse.append(y)
      sometin = []
      sometin.append(y)
    else:
      sometin.append(y)
  return sometinelse
```

```
def growth(x):
  alacart = []
  'establishes weekly average growth rates befoer and after'
  for y, z in enumerate(x['cases']):
    if y == 0:
      alacart.append(0)
      h = z
    else:
      a = (np.log(z) - np.log(h)) *100
      alacart.append(a)
      h = z
  averagesperweek = weekaverage(alacart)
  return averagesperweek
```
